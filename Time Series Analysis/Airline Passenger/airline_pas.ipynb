```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

#Loas Dataset
df = pd.read_csv('airline_passengers_dataset.csv',parse_dates=['Month'])
df.head(10)
```

```python
df
```

```python
df.index
```

```python
#Yearly Trends
%matplotlib inline
df.resample('Y').mean().plot()
```

```python
# Decompose the time series
result = seasonal_decompose(
    df['Passengers'],
    model='multiplicative',  # Airline dataset works best with multiplicative
    period=12                # 12 months = yearly seasonality
)

# Plot decomposition
result.plot()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

sns.lineplot(
    data=df,
    x='Month',
    y='Passengers'
)
```
