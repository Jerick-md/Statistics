```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample quarterly revenue data
df = pd.DataFrame(
    {'Revenue': [1200, 3444, 5667, 234, 4500, 5000, 5500, 6200]},
    index=pd.period_range('2025Q1', periods=8, freq='Q')
)

# 4-quarter moving average (1 year)
df['MA_4'] = df['Revenue'].rolling(window=4).mean()

# Plot
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(df.index.astype(str), df['Revenue'],
        marker='o', label='Revenue')
ax.plot(df.index.astype(str), df['MA_4'],
        marker='o', label='4-Quarter Moving Average')

ax.set_title('Quarterly Revenue and 4-Quarter Moving Average')
ax.set_xlabel('Quarter')
ax.set_ylabel('Revenue')
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
