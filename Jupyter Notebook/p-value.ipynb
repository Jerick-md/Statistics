```python
from scipy.stats import ttest_1samp
import numpy as np

#Quiz Score of student of QHSC
# Null Hypothesis = 3
#Alternatve Hypothesis != 3

score = [5,7,4,3,5,1,2,3,4,5,3]
```

```python
t_test, p_value = ttest_1samp(score,popmean=3) #3 is the Null Hypothesis


# P-Value
print(f'P-Value: {p_value}')
```

```python
'''
0.13 > 0.05, This conclude that We Didn't REJECT THE NULL HYPOTHESIS
'''
```
