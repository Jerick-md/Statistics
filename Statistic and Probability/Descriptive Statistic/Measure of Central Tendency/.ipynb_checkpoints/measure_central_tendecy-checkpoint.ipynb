MEASURE OF CENTRAL TENDENCY
 -Mahalaga ang Central Tendency (Mean, Median, Mode, Min, Max) nagbibigay ito ng isang representative value ng buong dataset. Sa halip na tingnan ang lahat ng data points isa-isa, makikita mo agad kung ano ang "karaniwang" value sa dataset

```python
import pandas as pd

df = pd.read_csv('students_exam_scores.csv')
df
```

```python
#Average Exam Score
print(f'Average exam score: {df['Exam_Score'].mean()}') 
```

```python
#Madalas na Score
print(f'Most frequent exam score: {df['Exam_Score'].mode()}')
```

```python
#Pinaka mababang Score
print(f'Minimum exam score: {df['Exam_Score'].min()}')
```

```python
#Pinaka Mataas na score
print(f'Maximum exam score: {df['Exam_Score'].max()}')
```
