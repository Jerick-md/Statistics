ANOVA


```python
import pandas as pd
from scipy.stats import f_oneway

'''
Null: Walang pagkakaiba sa avearge mean ng tatlong methods
Al_Null: Meron pagkakaiba sa average ng tatlong methods
'''
df = pd.read_csv('teaching_methods.csv')
df
```

```python
f_stat, p_value = f_oneway(df['Method_A'],df['Method_B'],df['Method_C'])

sig_lev = 0.05

if p_value < sig_lev:
    print('Meron Pag kakaiba sa tatlong Methods')
if p_value > sig_lev:
    print('Ang tatlong Methods ay walang Pagkakaiba')
    
print(f'F_stat: {f_stat}')
print(f'P_value: {p_value}') #0.0000000000000007083833097423345
```
